## 2 Card Black Jack Probability Distribution

- I will be looking at the distribution of pulling to cards in black Jack
- I will not be considering aces for the 3rd card draw
- Aces will either be 1 or 11 depending on if the player bust (over 21) 

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
class Deck:
    def __init__(self):
        cards = {
            'A': 4,
            1: 4,
            2: 4,
            3: 4,
            4: 4,
            5: 4,
            6: 4,
            7: 4,
            8: 4,
            10: 4, 
            'J': 4,
            'Q': 4,
            'K': 4
        }

        self.card_deck = []

        for card in cards.keys():
            for x in range(cards[card]):
                self.card_deck.append(card)

    
    def draw_random_card(self):
        random_index = random.randint(0, len(self.card_deck) - 1)
        return self.card_deck.pop(random_index)
```

```python
def calc_cards(card1, card2):
    ten_cards = ['J', 'Q', 'K']

    if card1 in ten_cards:
        card1 = 10

    if card2 in ten_cards:
        card2 = 10

    if card1 == 'A':
        card1 = 11

    if card2 == 'A':
        card2 = 11

    if card1 == 'A' and card2 == 'A':
        return 12

    return card1 + card2
```

```python
def play():
    play_deck = Deck()

    card1 = play_deck.draw_random_card()
    card2 = play_deck.draw_random_card()

    card_total = calc_cards(card1, card2)
    
    return card_total
```

```python
results = [play() for _ in range(1000)]

sns.histplot(results)
```
